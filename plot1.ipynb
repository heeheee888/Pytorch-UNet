```python
import matplotlib.pyplot as plt
import cv2
import os

path = "./test/result5/"
file_list = os.listdir(path)

#print ("file_list: {}".format(file_list))

for i, filename in enumerate(file_list):
    print(filename)
    imgg = cv2.imread(f'./test/mask/{filename[:-8]}.png')
    plt.subplot(3,2,1)
    plt.imshow(imgg)
    plt.title('mask', fontsize = 5)
    img = cv2.imread(f'./test/result4/{filename}')
    plt.subplot(3,2,3)
    plt.imshow(img)
    plt.title('MSE Loss + Dice Loss', fontsize = 5)
    img1 = cv2.imread(f'./test/result5/{filename}')
    plt.subplot(3,2,4)
    plt.imshow(img1)
    #plt.axis('off')
    plt.title('MSE Loss', fontsize = 5)
    img2 = cv2.imread(f'./test/result6/{filename}')
    plt.subplot(3,2,5)
    plt.imshow(img2)
    plt.title('L1 Loss + Dice Loss', fontsize = 5)
    img3 = cv2.imread(f'./test/result/{filename}')
    plt.subplot(3,2,6)
    plt.imshow(img3)
    plt.title('BCE Loss + Dice Loss', fontsize = 5)
    plt.tight_layout()
    #plt.show()
    plt.savefig(f'./plot/fig{i}.png', dpi = 150, facecolor = 'white')
```
